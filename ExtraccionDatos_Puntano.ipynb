```python
import requests
from google.colab import userdata
user_key=userdata.get('API_PAC_KEY')


url= 'https://www.presupuestoabierto.gob.ar/api/v1/pef'

headers = {"Authorization": user_key, 'Content-Type': 'application/json'}

pef = {
    "title": "Programación y ejecución física del primer trimestre",
    "columns": [
		"ejercicio_presupuestario",
		"trimestre",
		"jurisdiccion_id",
		"jurisdiccion_desc",
		"servicio_id",
		"servicio_desc",
		"programa_id",
		"programa_desc",
		"subprograma_id",
		"subprograma_desc",
		"tipo_medicion_fisica",
		"medicion_fisica_id",
		"medicion_fisica_desc",
		"unidad_medida_desc",
		"totalizador_avance_fisico",
		"ejecutado_vigente_trim",
		"ejecutado_acumulado_trim",
		"causa_desvio",
		"causa_desvio_detalle",
		"ultima_actualizacion_fecha"
    ],
    "ejercicios": [
        2023
    ],
	"filters": [
		{
		"column": "trimestre",
		"operator": "equal",
		"value": "2"
		}
	]
}
response = requests.post(url, headers=headers, json=pef)


response.text
import pandas as pd
from io import StringIO

pd.read_csv(StringIO(response.text))
```

```python
import PyPDF2

path='/content/PA_GuiaDatosAbiertos.pdf'
with open(path, 'rb') as f:
  pdf = PyPDF2.PdfReader(f)
  raw=""
  for page in range(len(pdf.pages)):
    raw+=pdf.pages[page].extract_text()
```
